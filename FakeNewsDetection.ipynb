# Importing Necessary Modules Needed

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

# Reading csv Data Files

```python
fake_data=pd.read_csv("Fake.csv")
true_data=pd.read_csv("True.csv")
```

```python
fake_data.head()
```

```python
true_data.head()
```

# Fetching The Data For Null Values

```python
print(fake_data.isnull().sum())
print(true_data.isnull().sum())
```

## Adding An Column That Represents Data (true,false) After Merging Both

```python
#Assigning '0' for false data and '1' for true data 
fake_data['class']=0
fake_data.head()
```

```python
true_data['class']=1
true_data.head()
```

## Removing unnecessary columns for avoiding noise in data

```python
fk=fake_data.drop(['title','subject','date'],axis=1)
fk.head()
```

```python
td=true_data.drop(['title','subject','date'],axis=1)
td.head()
```

```python
fk.shape,td.shape
```

## Removing some data from  dataset and store it for manual testing

```python
fake_data_for_manual_testing=fk.tail(10)
for i in range(23480,23470,-1):
    fk.drop([i],axis=0,inplace=True)
true_data_for_manual_testing=td.tail(10)
for i in range(21416,21406,-1):
    td.drop([i],axis=0,inplace=True)
```

```python
fake_data_for_manual_testing.head()
```

```python
true_data_for_manual_testing.head()
```

## Combining both the datasets into a one final dataset

```python
data=pd.concat([fk,td],axis=0)
data.shape
```

## Randomly Shuffling The DataFrame To Make Sure Model is Learning.

```python
data_final=data.sample(frac=1)
data_final.head(5)
```

```python
data_final.reset_index(inplace=True)
data_final.drop(['index'],axis=1,inplace=True)
```

```python
data_final.columns
```

```python
data_final.head(5)
```

## Creating A Function To Process The Text And Remove Noise In Text Data

```python
def purifynews(text):
    text=text.lower()
    text=re.sub(r'\[.*?\]','',text)
    text=re.sub(r"\\W"," ",text)
    text=re.sub(r'https?://\S+|www\.\S+','',text)
    text=re.sub(r'<.*?>+','',text)
    text=re.sub(r'[%s]' % re.escape(string.punctuation), '',text)
    text=re.sub(r'\n', '',text)
    text=re.sub(r'\w*\d\w*', '',text)
    return text
```

```python
data_final['text']=data_final['text'].apply(purifynews)
```

## Defining Independent And Dependent Variables

```python
x=data_final['text']
y=data_final['class']
```

## Spliting train and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

## Convert Text To Vectors for model underrstanding

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
xv_train=vectorizer.fit_transform(x_train)
xv_test=vectorizer.transform(x_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_logistic_regresion=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_logistic_regresion))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt=DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

## Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state=0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

## Model Testing

```python
def output_decide(n):
    if n==0:
        return "Fake News"
    else:
        return "Not A Fake News"
def manual_testing(text):
    testing_news={"text":[text]}
    new_test_data=pd.DataFrame(testing_news)
    new_x_test=new_test_data["text"].apply(purifynews)
    new_xv_data=vectorizer.transform(new_x_test)
    pred_lr=LR.predict(new_xv_data)
    pred_dt=DT.predict(new_xv_data)
    pred_rf=RF.predict(new_xv_data)
    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nRFC Prediction: {}".format(output_decide(pred_lr[0]),
                                                                                                              output_decide(pred_dt[0]), 
                                                                                                              output_decide(pred_rf[0])))
```
